```python
class UserATM():
     
    def __init__(self, name, account_number):
        self.name = name
        self.account_number = account_number
        self.balance = 0
        self.pin = 1234
        
    def my_details(self):
        print("*Account Details*")
        print(f"Your Account Name: {self.name}")
        print(f"Your Account Number: {self.account_number}")
        print(f"Balance: N {self.balance}")
        
    def check_balance(self):
        print(f"Your available balance is: N{self.balance}\n")
        
    def deposit_cash(self, deposit_cash):
        self.balance = self.balance + deposit_cash
        # self.balance += deposit_cash
        print(f"Your amount {deposit_cash} is successful\n Available balance is {self.balance}")
        
    def withdraw_cash(self, withdraw_cash):
        if withdraw_cash > self.balance:
            print(f"You have insufficient balance.\n Available balance is {self.balance}")
        else:
            self.balance = self.balance - withdraw_cash
            # self.balance -= withdraw_cash
            print(f"The amount {withdraw_cash} is successful\n Available balance is {self.balance}")
        
    def next_trans (self, response):
        self.response = response
        
#         print("""  Press Y to continue or N to exit: Y. Continue
#                                                      N. Exit  """)

        
        if self.response == 'Y':
            self.UserATM_menu()
        else:
            print("Thank you dear customer for banking with us.")
        
        
    def UserATM_menu(self):
        
        print("""
               
            MENU:
            0. Exit
            1. Check Balance
            2. Deposit
            3. Withdraw
            4. Account details
        
        """)
        
        while True:
            
            number = int(input('Enter a number from the menu '))
            
            if number == 0:
                print('Exit now')
                break
                
            elif number in (1,2,3,4):
                maximum_login = 3
                
                while(maximum_login != 0):
                    pin = int(input('Welcome, Enter your pin '))
                    
                    if pin == self.pin:
                        print("Welcome to Access Bank Plc.")
                        
                        if number == 1:
                            self.check_balance()
                            print("Transaction Successful. Do you want to perform another transaction?")
                            response = input('Press Y to continue or N to exit: ')
                            self.next_trans(response)
                            break
                            
                        elif number == 2:
                            amount = int(input('Enter cash deposit amount: '))
                            self.deposit_cash(amount)
                            print("Transaction Successful. Do you want to perform another transaction?")
                            response = input('Press Y to continue or N to exit: ')
                            self.next_trans(response)
                            break
                            
                        elif number == 3:
                            amount = int(input('Enter cash withdrawal amount: '))
                            self.withdraw_cash(amount)
                            print("Transaction Successful. Do you want to perform another transaction?")
                            response = input('Press Y to continue or N to exit: ')
                            self.next_trans(response)
                            break
                            
                        elif number == 4:
                            self.my_details()
                            print("Transaction Successful. Do you want to perform another transaction?")
                            response = input('Press Y to continue or N to exit: ')
                            self.next_trans(response)
                    
                    else:
                        maximum_login -= 1
                        print(f'Wrong pin, you have {maximum_login} attempts left\n')
                        
                        if maximum_login == 0:
                            print("Invalid input!")
                            print("\t\t0. Enter 0 to Logout and Exit!")
                            False
```
